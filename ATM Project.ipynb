## ATM Project

```python
class atm:
    def __init__(self,balance=0):
        while True:
            self.name=input("enter your name")
            if self.name.replace(" ",'').isalpha():
                break
            else:
                print(" Name should be contain alphabate charector only: Try Again")
        while True:
            self.accountNO=input("enter your account number")
            if self.accountNO.isnumeric() and len(self.accountNO)==16:
                break
            else:
                print(" enter a valid Account number ")
        
        while True:
            self.PIN=input("enter your pin")
            if self.PIN.isnumeric() and len(self.PIN)==4:
                break
            else:
                print(" enter a valid PIN and PIN length must be 4 digit")
        self.balance=balance

    def details(self):
        ''' display details for person'''
        pin=input("enter your pin")
        if pin==self.PIN:
            print(f" Account Holder name is {self.name}")
            print(f" Account number is {self.accountNO[0:4]} **** **** {self.accountNO[-4:16]}")
            print(f" Current balance is {self.balance}")
        else:
            print("enter correct pin")
    def displayBalance(self):
        pin=input("enter your pin")
        if pin==self.PIN:
            print(f"your current balance is {self.balance} ")
        else:
            print("enter correct pin")
    def depositeBalance(self,depo):
        if depo>0:
            pin=input("enter a pin")
            if pin==self.PIN:
                self.balance+=depo
                print(f" balance deposite successfully \n now your current blance is : {self.balance} ")
            else:
                print(" enter a correct PIN")
        else: 
            print(" deposite amount must be greater then  0")
    def withdrawBalance(self,amount):
        if self.balance>=amount:
            pin=input("enter a current PIN")
            if pin==self.PIN:
                self.balance-=amount
                print(f" balance withdraw successfully \n your current balance is {self.balance}")
            else:
                print(" enter a correct PIN")
        else:
            print("Insufficient Balance")
    def ChangeName(self,NewName):
        pin=input("enter your current PIN")
        if pin==self.PIN:
            if self.name!=NewName:
                self.name=NewName
                print(f" your Account Holder name change successfully \n your name is {self.name}")
            else:
                print(" new name and old name can not be same ")
        else:
            print("enter a valid PIN")
    def showaccountNO(self):
        pin=input("enter your PIN")
        if pin==self.PIN:
            print(f" your account number is : {self.accountNO}")
        else:
            self.showpin()
    def showpin(self):
        choise=input("would you like to see your pin ? (yes/no)")
        if choise in ["yes","YES","Yes","y","Y"]:
            enter_name=input("enter your name")
            if enter_name==self.name:
                print(f" your current PIN is : {self.PIN}")
            else:
                print(" enter a valid name ")
        else:
            print(" Return to menu")
    def changePIN(self,oldpin,newpin=0,conformPIN=0):
        if oldpin==self.PIN:
            newpin=input(" enter a new PIN")
            conformPIN=input("enter a conform PIN")
            if newpin==conformPIN:
                if len(newpin)==4:
                    if newpin==oldpin:
                        print(" new PIN and old PIN can not be same ")
                    else:
                        self.PIN=newpin
                        print(" your PIN updated successfully")
                else:
                    print(" enter valid PIN")
            else:
                print(" enter a same PIN in conformPIN and newpin")
        else:
            print(" enter a valid current PIN")
            
        

    def trans(self):
        print(""" 
        \n1. Account Details
        \n2. Account Balance
        \n3. Deposite Balance
        \n4. Withdraw Balance
        \n5. Change Name
        \n6. ShowAccount Number
        \n7. change PIN
        \n8. Exit""")
        while True:
            try:
                opt=int(input("choose option"))
            except:
                print(" choose correct option")
                continue
            else:
                if opt==1:
                    atm.details(self)
                       
                elif opt==2:
                    atm.displayBalance(self)
            
                   
                elif opt==3:
                    depo=int(input("enter your ammount"))
                    atm.depositeBalance(self,depo)
                elif opt==4:
                    amount=int(input("enter your amount"))
                    atm.withdrawBalance(self,amount)       
                elif opt==5:
                    NewName=input("enter your new name")
                    atm.ChangeName(self,NewName)
                elif opt==6:
                    atm.showaccountNO(self)
                elif opt==7:
                    oldpin=input("Enter your current PIN: ")
                    atm.changePIN(self,oldpin)
                elif opt==8:
                    print(" thanks for using ATM")
                    break
obj=atm()
obj.trans()
            
```
